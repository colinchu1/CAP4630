#Imports

```
from mpl_toolkits import mplot3d
```

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

##Function 1 (minimum)
$$({2x+2)^2}+{y^2}+20$$

Display 3D PLOT

```
def f(x, y):
    return (2*x+2) ** 2 + y ** 2 + 20;

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
zmin = np.min(Z)
print(zmin);
xmin, ymin = np.unravel_index(np.argmin(Z), Z.shape)
xmax, ymax = np.unravel_index(np.argmax(Z), Z.shape)
mi = (X[xmin,ymin], Y[xmin,ymin], Z.min())
ma = (X[xmax,ymax], Y[xmax,ymax], Z.max())
Ami = np.array([mi])
Ama = np.array([ma])
print(Ami)
ax.plot(Ami[:,0], Ami[:,1], Ami[:,2], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

###Density Plot(minimum)

```
zmin = np.min(Z)
print(zmin);
plt.plot(Ami[:,0], Ami[:,1],  markerfacecolor = 'white', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
# Change color palette
plt.pcolormesh(X, Y, Z)
plt.colorbar();
plt.show()
```

###Countour Plot(minimum)

```
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
plt.plot(Ami[:,0], Ami[:,1],  markerfacecolor = 'white', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

#Function 2, 3D plot(maximum)
$$-(({3x+2)^3}+{3y^4})$$

```
def f(x, y):
    return -(3*x **3 + 3* y**4)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
zmax = np.max(Z)
print(zmax);
xmax, ymax = np.unravel_index(np.argmax(Z), Z.shape)
ma = (X[xmax, ymax], Y[xmax, ymax], Z.max())
Ama = np.array([ma])
ax.plot(Ama[:,0], Ama[:,1], Ama[:,2], markerfacecolor = 'white', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

##Maximum density plot

```
plt.plot(Ama[:,0], Ama[:,1],  markerfacecolor = 'red', markeredgecolor = 'black', marker = 'o', markersize = 10, alpha = 1, zorder = 100)
# Change color palette
plt.pcolormesh(X, Y, Z)
plt.colorbar();
plt.show()
```

##contour plot(maximum)

```
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar();
plt.plot(Ama[:,0], Ama[:,1],  markerfacecolor = 'red', markeredgecolor = 'black', marker = 'o', markersize = 10, alpha = 1, zorder = 100)
```

#Function 3 (zero gradient not max nor min)
$${2x^2}-{5*y^2}$$

```
def f(x, y):
    return (2*x**2) - 5*y**2; 

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
zmin = np.min(Z)
print(zmin);
xmin, ymin = np.unravel_index(np.argmin(Z), Z.shape)
xmax, ymax = np.unravel_index(np.argmax(Z), Z.shape)
mi = (X[xmin,ymin], Y[xmin,ymin], Z.min())
Ami = np.array([mi])
print(Ami)
ax.plot([0], [0],[0], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```
